```python
#matrix of user_id and rattings
moviem = df.pivot_table(index= 'user_id',columns = 'title',values = 'rating')
moviem
```

```python
starwars = moviem["Star Wars (1977)"]
starwars
```

```python
moviem.corrwith(starwars) #gives the corelation with
#if a certain user didnot ratted both then it will give nan
```

```python
starwarsdf = pd.DataFrame(moviem.corrwith(starwars),columns = ['Correlation'])
starwarsdf.dropna(inplace = True)
```

```python
starwarsdf.sort_values('Correlation',ascending = False)
```

```python
#join with no of rattings
starwarsdf = starwarsdf.join(rattings['No Of rattings'])
```

```python
starwarsdf
```

```python
#eliminate the movies that have  no of rattings less than 100
starwarsdf[starwarsdf['No Of rattings']>100].sort_values(by = 'Correlation',ascending = False)
```

## Predict Function

```python
def predict_movies(movie):
    movie_info = moviem[movie] #get the row from movie m matrix
    simillar_movie = moviem.corrwith(movie_info) #calculate the correlation with other matrix
    
    movie_df = pd.DataFrame(simillar_movie,columns = ['Correlation']) # convert it to data frame
    movie_df.dropna(inplace = True) # drop nan value
    
    corr_movie = movie_df.join(rattings['No Of rattings']) #join with the no of rattings from rattings dataframe
    
    prediction = corr_movie[corr_movie['No Of rattings']>100].sort_values(by = 'Correlation',ascending = False)
    #get the correlated movies which have the minimum ratting of 100
    return prediction #return the df of predicted movie


    
    
```

```python
predictions = predict_movies("Star Wars (1977)")
```

```python
predictions.head()
```

```python
predictions = predict_movies("Titanic (1997)")
```
